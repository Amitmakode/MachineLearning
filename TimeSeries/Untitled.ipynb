```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
data=pd.read_csv("TSLA.csv")
```

```python
data.info()
```

```python
#simple moving avg
```

```python
data["Open"].plot(figsize= (12, 4))
```

```python
data["Open: 10 days rolling"] = data["Open"].rolling(window=5, min_periods=1 ).mean()
```

```python
data[["Open","Open: 10 days rolling"]].plot(figsize=(15,6))
```

```python
data["Open: 30 days rolling"] = data["Open"].rolling(window=30, min_periods=1 ).mean()
data["Open: 50 days rolling"] = data["Open"].rolling(window=50, min_periods=1 ).mean()
```

```python
data[["Open","Open: 10 days rolling","Open: 30 days rolling","Open: 50 days rolling"]].plot(figsize=(15,6))
```

```python
## cummilative moving avg
data['Open'].expanding().mean().plot(figsize=(10,5))
```

```python
## exponenetial moving avg
## lets smoothing factor is = 0.1
data['EMA_0.1']=data['Open'].ewm(alpha=0.1, adjust=False).mean()
data['EMA_0.3']=data['Open'].ewm(alpha=0.3, adjust=False).mean()
```

```python
data[["Open","EMA_0.1"]].plot(figsize=(15,6))
```

```python
data[["Open","EMA_0.1","EMA_0.3"]].plot(figsize=(15,6))
```

```python
data['EWM 5day']=data['Open'].ewm(span= 5).mean()
```

```python
data[["Open","EMA_0.1","EMA_0.3",'EWM 5day']].plot(figsize=(15,6))
```
